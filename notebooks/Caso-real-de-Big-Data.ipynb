# Caso Real de Big Data - Spotify

### **Que datos genera?**
* **Tipo:** Datos como historial de reproduccion, likes, ubicacion, y datos no estructurados como archivos de audio e imagenes de portadas.
* **Volumen:** Almacena y procesa petabytes de informacion provenientes de un listado de mas de 100 millones de canciones y las interacciones de millones de usuarios activos.
* **Frecuencia:** Transmision continua de datos en tiempo real generada por cada reproduccion, pausa, salto de pista o busqueda a nivel global.

### **Quien los usa?**
* **Spotify:** Sus equipos de ingenieria entrenan modelos de machine learning para el motor de recomendacion.
* **Artistas y Discograficas:** Acceden a traves de "Spotify for Artists" para analizar la demografia de sus oyentes y planificar giras segun la demanda local.
* **Anunciantes:** Consumen datos agregados y segmentados para dirigir campanas publicitarias personalizadas.

### **Que problema resuelve?**
* **Valor:** Soluciona la saturacion de contenido mediante la personalizacion como "Discover Weekly" o "Wrapped", lo que incrementa la retencion de usuarios, optimiza la distribucion de servidores y conecta de forma eficiente a los creadores con su audiencia ideal.

### **Por que no bastaria Excel?**
* **Volumen inviable:** Excel tiene un limite fisico de 1,048,576 filas, los eventos globales de Spotify superan este numero en cuestion de minutos.
* **Velocidad insostenible:** Las hojas de calculo o bases de datos relacionales simples se colapsarian al intentar procesar e indexar millones de flujos simultaneos en tiempo real.
* **Complejidad:** Excel carece del entorno de computacion distribuida necesario para ejecutar los algoritmos de filtrado colaborativo y redes neuronales que procesan el comportamiento de millones de personas a la vez.
