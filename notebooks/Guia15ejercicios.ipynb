```python
#EJERCICIO 1
def clasificar_edad(edad):
    if edad < 13:
        return "Niño"
    elif edad <= 17:
        return "Adolescente"
    elif edad <= 59:
        return "Adulto"
    else:
        return "Adulto mayor"

edad = 21
categoria = clasificar_edad(edad)
print(f"Edad: {edad} años")
print(f"Categoría: {categoria}")
```

```python
#EJERCICIO 2
def convertir_a_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

celsius = 25
fahrenheit = convertir_a_fahrenheit(celsius)
print(f"Temperatura en Celsius: {celsius}°C")
print(f"Temperatura en Fahrenheit: {fahrenheit}°F")
```

```python
#EJERCICIO 3
def calculadora(num1, num2, operacion):
    if operacion == "suma":
        return num1 + num2
    elif operacion == "resta":
        return num1 - num2
    elif operacion == "multiplicacion":
        return num1 * num2
    elif operacion == "division":
        if num2 != 0:
            return num1 / num2
        else:
            return "Error: No se puede dividir entre cero"
    else:
        return "Error: Operación no válida"

num1 = 10
num2 = 5
operacion = "division"

resultado = calculadora(num1, num2, operacion)
print(f"Números: {num1} y {num2}")
print(f"Operación: {operacion}")
print(f"Resultado: {resultado}")
```

```python
# EJERCICIO 6: Contador de vocales

def contar_vocales(frase):
    vocales = "aeiou"
    contador = 0

    for letra in frase.lower():
        if letra in vocales:
            contador += 1

    return contador

frase = "La programacion en Python es muy util"
cantidad_vocales = contar_vocales(frase)

print(f"Frase: {frase}")
print(f"Cantidad de vocales: {cantidad_vocales}")
```

## Ejercicio 10: Diccionario de estudiante
**Enunciado:** Crear un diccionario que almacene nombre, edad, carrera, promedio y estado. Mostrar los datos en formato ordenado y explicar cómo se accede a los valores.

```python
estudiante = {
    "nombre": "Esly Díaz",
    "edad": 22,
    "carrera": "Ingeniería en Sistemas",
    "promedio": 88.5,
    "estado": "Activo"
}

print("=== Datos del Estudiante ===")
for campo, valor in estudiante.items():
    print(f"  {campo.capitalize()}: {valor}")
```

**Explicación:** Un diccionario en Python almacena pares clave-valor definidos entre llaves `{}`. Se accede a un valor individual usando `diccionario["clave"]`. Para recorrer todos los pares a la vez se utiliza `.items()`, que retorna cada par como una tupla `(clave, valor)`. El método `.capitalize()` pone en mayúscula la primera letra de cada campo para una presentación más ordenada.

## Ejercicio 11: Función para calcular descuento
**Enunciado:** Crear una función llamada `calcular_descuento` que reciba el precio original y el porcentaje de descuento, y retorne el precio final.

```python
def calcular_descuento(precio_original, porcentaje_descuento):
    monto_descuento = precio_original * (porcentaje_descuento / 100)
    precio_final = precio_original - monto_descuento
    return precio_final

precio = 1500.00
descuento = 20

precio_con_descuento = calcular_descuento(precio, descuento)
print(f"Precio original:    L. {precio:.2f}")
print(f"Descuento aplicado: {descuento}%")
print(f"Precio final:       L. {precio_con_descuento:.2f}")
```

**Explicación:**  
- **Entrada:** `precio_original` (número con decimales) y `porcentaje_descuento` (valor entre 0 y 100).  
- **Proceso:** Se calcula el monto del descuento multiplicando el precio por el porcentaje dividido entre 100, luego se resta al precio original.  
- **Salida:** La función retorna el `precio_final` ya con el descuento aplicado. Los valores están definidos directamente en el código; no se solicita ningún dato al usuario.

## Ejercicio 12: Validación de contraseña
**Enunciado:** Validar una contraseña definida en el código. Debe tener al menos 8 caracteres, un número, una letra mayúscula y una letra minúscula.

```python
def validar_contrasena(contrasena):
    errores = []

    if len(contrasena) < 8:
        errores.append("Debe tener al menos 8 caracteres.")
    if not any(c.isdigit() for c in contrasena):
        errores.append("Debe contener al menos un número.")
    if not any(c.isupper() for c in contrasena):
        errores.append("Debe contener al menos una letra mayúscula.")
    if not any(c.islower() for c in contrasena):
        errores.append("Debe contener al menos una letra minúscula.")

    if not errores:
        return True, "Contraseña válida."
    else:
        return False, errores

# Contraseña definida directamente en el código (sin input del usuario)
contrasena = "Python2024"

valida, mensaje = validar_contrasena(contrasena)
print(f"Contraseña evaluada: '{contrasena}'")
print(f"¿Es válida?: {valida}")
if isinstance(mensaje, list):
    print("Errores encontrados:")
    for error in mensaje:
        print(f"  - {error}")
else:
    print(f"Resultado: {mensaje}")
```

**Explicación:** La contraseña está definida directamente como variable en el código. La función `validar_contrasena` evalúa cuatro condiciones usando `any()` con expresiones generadoras:  
- `.isdigit()`: verifica si un caracter es un dígito numérico.  
- `.isupper()`: verifica si un caracter es letra mayúscula.  
- `.islower()`: verifica si un caracter es letra minúscula.  

Los errores se acumulan en una lista. La función retorna una tupla `(booleano, mensaje_o_lista)`. Se usa `isinstance()` para determinar si el mensaje es una lista de errores o un string de éxito, y mostrar la salida correspondiente.

```python
# Ejercicio 13: Sistema simple de inventario

# Enunciado:
# Crear una lista de productos con código, nombre, precio y cantidad disponible.
# Mostrar productos, buscar uno por código y calcular el valor total del inventario.

# Lógica:
# Se utiliza una lista de diccionarios para guardar productos.
# Luego se recorren con ciclos for para mostrar información y calcular resultados.

productos = [
    {"codigo": "P001", "nombre": "Laptop", "precio": 15000, "cantidad": 3},
    {"codigo": "P002", "nombre": "Mouse", "precio": 250, "cantidad": 10},
    {"codigo": "P003", "nombre": "Teclado", "precio": 600, "cantidad": 5}
]

print("Lista de productos:")

for producto in productos:
    print(producto)

codigo_buscar = "P002"
producto_encontrado = None

for producto in productos:
    if producto["codigo"] == codigo_buscar:
        producto_encontrado = producto

print("\nProducto encontrado:")
print(producto_encontrado)

valor_total = 0

for producto in productos:
    valor_total += producto["precio"] * producto["cantidad"]

print("\nValor total del inventario:", valor_total)
```

```python
# Ejercicio 14: Registro de ventas

# Enunciado:
# Crear una lista de ventas con cliente, producto, cantidad y precio unitario.
# Calcular el total por venta, total general vendido
# e identificar el cliente con la compra más alta.

# Lógica:
# Se utiliza una lista de diccionarios para almacenar las ventas.
# Se recorren los datos con un ciclo for para calcular totales
# y determinar la compra más alta.

ventas = [
    {"cliente": "Carlos", "producto": "Laptop", "cantidad": 1, "precio_unitario": 15000},
    {"cliente": "María", "producto": "Mouse", "cantidad": 3, "precio_unitario": 250},
    {"cliente": "José", "producto": "Teclado", "cantidad": 2, "precio_unitario": 600}
]

total_general = 0
compra_mas_alta = 0
cliente_compra_mas_alta = ""

print("Registro de ventas:\n")

for venta in ventas:
    
    total_venta = venta["cantidad"] * venta["precio_unitario"]
    
    print("Cliente:", venta["cliente"])
    print("Producto:", venta["producto"])
    print("Cantidad:", venta["cantidad"])
    print("Precio unitario:", venta["precio_unitario"])
    print("Total venta:", total_venta)
    print("--------------------------")
    
    total_general += total_venta
    
    if total_venta > compra_mas_alta:
        compra_mas_alta = total_venta
        cliente_compra_mas_alta = venta["cliente"]

print("\nTotal general vendido:", total_general)
print("Cliente con la compra más alta:", cliente_compra_mas_alta)
print("Monto de compra más alta:", compra_mas_alta)
```

```python
# Ejercicio 15: Limpieza básica de texto

# Enunciado:
# Limpiar una frase eliminando espacios innecesarios,
# convertir el texto a minúsculas y contar palabras.

# Lógica:
# Se utilizan métodos de cadenas como strip(), lower(),
# replace() y split() para limpiar y analizar el texto.

frase = "   Hola   jasiel,   este   es   un   ejercicio   de   Python   "

print("Frase original:")
print(frase)

frase_limpia = frase.strip()

frase_limpia = frase_limpia.lower()

while "  " in frase_limpia:
    frase_limpia = frase_limpia.replace("  ", " ")

palabras = frase_limpia.split()

cantidad_palabras = len(palabras)

print("\nFrase limpia:")
print(frase_limpia)

print("\nCantidad de palabras:")
print(cantidad_palabras)

print("\nExplicación de métodos:")
print("strip(): elimina espacios al inicio y al final.")
print("lower(): convierte el texto a minúsculas.")
print("replace(): reemplaza texto.")
print("split(): separa el texto en palabras.")
```
